```python
import torch 
import torch.nn as nn

import seaborn as sns
```

```python
nn.init.calculate_gain(nonlinearity='leaky_relu', param= 0.2)
```

```python
w = torch.rand(64,3,300, 300)
```

```python

sns.kdeplot(nn.init.uniform_(w).flatten().numpy());
```

```python
sns.kdeplot(nn.init.normal_(w).flatten().numpy());
```

```python
sns.kdeplot(nn.init.sparse_(w).flatten().numpy());
```

```python
w = torch.empty(3, 16, 5, 5)
nn.init.dirac_(w)
w = torch.empty(3, 24, 5, 5)
nn.init.dirac_(w, 3)
```
