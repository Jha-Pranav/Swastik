Campesterol, stigmasterol, and β-sitosterol are plant sterols, which are similar in structure to cholesterol. These plant sterols can compete with cholesterol for absorption in the gut, potentially reducing cholesterol levels in the blood.

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np



# Calculate the WCSS for different values of k
wcss = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(data.values)
    wcss.append(kmeans.inertia_)

# Plot the results
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```
